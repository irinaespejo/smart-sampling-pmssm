```python
import numpy as np
np.random.seed(123)
import matplotlib.pyplot as plt
import skopt
from skopt.sampler import Lhs
from skopt.sampler import Grid
from scipy.spatial.distance import pdist
```

```python
space = [
   skopt.space.Real(50., 5000., name='mzp', prior='uniform'),
   skopt.space.Real(100., 1200., name='mdm', prior='uniform'),
   skopt.space.Real(50., 170., name='mdh', prior='uniform'),
   skopt.space.Real(0.1, 3.5, name='g', prior='log-uniform')]

lhs = Lhs(lhs_type="centered", criterion=None)

n_samples = 1000

x = np.array(lhs.generate(space, n_samples, random_state=42))
```

```python
x.shape
```

```python
import pandas as pd
data = pd.DataFrame(data = x, columns = ["mzp","mdm", "mdh", "g"])
data.to_csv('./hypercube_param_samples.csv')
```
